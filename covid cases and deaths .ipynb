```python
vac.loc['date'] = pd.to_datetime(vac['date'])
vac = vac[vac['date'].dt.year == 2024]
vac
```

```python
print(df['date'].unique())
```

```python
df.loc['date'] = pd.to_datetime(df['date'])
df_2024 = df[df['date'].dt.year == 2024].copy()
df_2024
```

```python
merged_df = pd.merge(df_2024, vac, how = "inner", on=['country', 'date'])
merged_df
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(
    merged_df['total_vaccinations'],
    merged_df['new_cases'])
```

```python
import matplotlib.dates as mdates
plt.tick_params(axis='both', which='major', labelsize=6) 
plt.title('Global COVID-19 Trends: Average of New Cases, Vaccinations, New Deaths')
merged_df['new_cases_avg'] = merged_df['new_cases'].rolling(window=1).mean()
plt.plot(merged_df['date'],merged_df['new_cases_avg'],label = 'New Cases', color = 'red', )
plt.xticks(rotation=45, ha='right')
plt.plot(merged_df['date'],merged_df['daily_vaccinations'], label = 'New Vaccinations', color = 'blue', )
plt.plot(merged_df['date'], merged_df['new_deaths'], label = "New Deaths", color = 'green',)
plt.xlabel("Date",fontsize=14)
ax = plt.gca()
ax.xaxis.set_major_locator(mdates.DayLocator(interval=4))
ax.xaxis.set_minor_locator(mdates.DayLocator(interval=2))
plt.ylabel("Frequency")
plt.legend()
plt.ylim(0, 4000)
```
